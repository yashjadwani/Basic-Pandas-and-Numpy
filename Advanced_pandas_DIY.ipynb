```python
# Q17: group student by their hobby and their course type
stds_gb = stds.groupby(['student_hobbies','student_course']).groups
stds_gb
```

```python
# Q18: find the standard deviation of the marks for students whose hobby is Sports

#df_gb['employee_salary'].agg([np.mean,np.sum,np.std,np.max,np.min]).round(2)

stds_gb = stds.groupby('student_hobbies')
stds_1 = stds_gb['student_marks'].agg([np.mean,np.min,np.max,np.std])
stds_1.loc['Sports']
```

```python
for name,dtls in stds_gb:
    if name =='Sports':
        print(dtls['student_marks'].agg([np.mean,np.min,np.max,np.std]).round(2))
```

# Pivot, Concat and Stack etc.

Download the two excel files from here

```python
# Q19: read the two csv files
df1 = pd.read_excel('D1-P1.xlsx')
```

```python
# Q20: read the two csv files
df2 = pd.read_excel('D1-P2.xlsx')
```

```python
# Q21: concat them into a single dataframe (outer join)
result = pd.concat([df1,df2],join='outer')
result
```

```python
# Q22: concat them into a single dataframe (inner join)
result = pd.concat([df1,df2],join='inner')
result
```

## Pivoting

```python
# Q23: pivot using city column as index and column set to date
r1=result.pivot_table(index = 'City', columns ='date',values ='Temperature')
r1
```

```python
# Q24: add a column with the mean temperature for each city over the 5 days
result.pivot_table(index = 'City', columns ='date', values='Temperature',margins=True)
```

```python
# Q25: Use melt to get back the original dataframe structure from the pivoted one
# City, date and temperature
# you can sort the city values in ascending form
f1= r1.melt(ignore_index=False, value_name='Temperature')
#f1 =f1.reset_index()
f1 = f1.sort_values(by='City')
f1
```
